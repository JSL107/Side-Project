```python
import os           #파일접근
import numpy as np              # 다차원 배열, 수학함수
import matplotlib.pyplot as plt   #데이터 시각화 = 그래프 패키지
import pandas as pd     #데이터를 저장하고 처리하는 패키지
%matplotlib inline

import tensorflow as tf
from tensorflow.keras import datasets
from tensorflow.keras import layers
from tensorflow.keras import models

from tensorflow.keras.models import load_model

```

```python
# 사용하는 모델의 label 맞춰주기
labels=[
    '내추럴',
    '로멘틱',
    '모던',
    '캐주얼',
    '클래식',
    '페미닌'
]
```

```python
# Error tokenizing data. C
# 구분자 명시
product_img = pd.read_csv('csv/kfashion.csv', encoding='cp949',sep='\t')
```

```python
product_img = np.array(product_img)
```

```python
product_img_arr=product_img[2][0].split(',')
# 이미지 url
product_img_arr[4]
```

```python
import requests
from io import BytesIO
from PIL import Image
```

```python
# 모델 불러오기
model = load_model('model/keras_model7000-6-choice.h5')
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='sgd', 
              metrics=['accuracy'])
```

```python
csvdata = []
pass1=[1500,2774,2970,4939] # 오류나는 컬럼 패스
# range 10014
for i in range(10014):
    product_img_arr=product_img[i][0].split(',')
    
    # 다운받을 이미지 url
    url = product_img_arr[4]
    # request.get 요청
    res = requests.get(url)
    #Img open
    image = Image.open(BytesIO(res.content))
    image = image.resize((224,224),resample=0)
    image = np.array(image)
    #image_arr = image.reshape(1,224,224,3)
    if i in pass1 : 
        continue
    image = image.reshape((1, image.shape[0],image.shape[1],image.shape[2]))

    # 모델 사용하기
    output = model.predict(image)
    product_img_arr.append(labels[np.argmax(output)])
    # csv에 값 추가하기
    csvdata.append(product_img_arr)
    #print(labels[np.argmax(output)])
    print(i)
  
```

```python
df = pd.DataFrame(csvdata)
df.to_csv('csv/스타일추가된 컬러테이블.csv', index=False, encoding='cp949') 
```
